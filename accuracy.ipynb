```python
import torch
from pathlib import Path

path = Path("E:/Computer Vision/checkpoints/khana_best.pt")
ckpt = torch.load(path, map_location="cpu", weights_only=False)

state = ckpt.get("model_state", ckpt.get("model", {}))
total_params = sum(v.numel() for v in state.values() if hasattr(v, "numel"))
file_mb = path.stat().st_size / 1e6

print("=" * 45)
print("  CHECKPOINT SUMMARY")
print("=" * 45)
print(f"  file         : {path.name}")
print(f"  size         : {file_mb:.1f} MB")
print(f"  arch         : {ckpt.get('arch', 'n/a')}")
print(f"  img_size     : {ckpt.get('img_size', 'n/a')} px")
print(f"  num_classes  : {len(ckpt.get('classes', []))}")
print(f"  total params : {total_params/1e6:.1f}M")
print(f"  val_acc1     : {ckpt.get('val_acc1', 0.0):.4f}")
print(f"  epoch        : {ckpt.get('epoch', 'not saved')}")
print(f"  cosine_clf   : {ckpt.get('cosine_clf', 'n/a')}")
print("=" * 45)
print("  CLASSES (first 10)")
print("=" * 45)
for c in ckpt.get('classes', [])[:10]:
    print(f"  {c}")
print(f"  ... ({len(ckpt.get('classes', []))} total)")
print("=" * 45)
print("  ALL SAVED KEYS")
print("=" * 45)
for k in ckpt.keys():
    print(f"  {k}")
```
